```python
#爬图图
import urllib.request
import re
import os
import urllib
import os

def get_html(url):#获取网页
    page = urllib.request.urlopen(url)#打开网址
    html_a = page.read()
    return html_a.decode('utf-8')#把网页按其编码方式解码，查询网页编码：document.charset


def get_img(html):#下载图片
    reg = r'https://[^\s]*?\.jpg'
    imgre = re.compile(reg)  # 转换成一个正则对象
    imglist = imgre.findall(html)  # 表示在整个网页过滤出所有图片的地址，放在imgList中
    x = 0        # 声明一个变量赋值,表示图片数
    path = 'D:\\Anaconda\\pycode\\images'  # 设置图片的保存地址
    if not os.path.isdir(path):
        os.makedirs(path)  # 判断没有此路径则创建
    paths = path + '\\'  # 保存在test路径下
    for imgurl in imglist:
        urllib.request.urlretrieve(imgurl, '{0}{1}.jpg'.format(paths, x))  # 打开imgList,下载图片到本地
        print(x)
        x = x + 1
    return imglist

def delete_img(path, img_name):#删图
    for root,dirs,files in os.walk(path):
        for name in files:
            if name==img_name+".jpg":
                os.remove(os.path.join(root,name))#删去该图
                print("delete success!")
            else:
                print("fail to find!")#不存在该图
                break
                
html_b = get_html("https://image.baidu.com/search/index?tn=baiduimage&ipn=r&ct=201326592&cl=2&lm=-1&st=-1&fm=result&fr=&sf=1&fmq=1595683463181_R&pv=&ic=&nc=1&z=&hd=&latest=&copyright=&se=1&showtab=0&fb=0&width=&height=&face=0&istype=2&ie=utf-8&sid=&word=%E6%97%A0%E6%B0%B4%E5%8D%B0%E6%BC%AB%E5%A8%81%E4%BA%BA%E7%89%A9")  # 获取该网页的详细信息
get_img(html_b)  # 从网页源代码中分析下载保存图片
```

```python
#为png图加水印
from PIL import Image, ImageDraw, ImageFont
import random
path="D:\\Anaconda\\pycode\\images_png\\"
savePath1="D:\\Anaconda\\pycode\\images_center"
if not os.path.isdir(savePath1):
    os.makedirs(savePath1)
savePath2="D:\\Anaconda\\pycode\\images_random"
if not os.path.isdir(savePath2):
    os.makedirs(savePath2)
        
for i in range(1,151):
    image=Image.open(path+str(i)+".png")
    image=image.resize((500,500),Image.ANTIALIAS)
    layer=image.convert('RGBA')#把图像转成RGB图
    text='watermark'#水印内容
    text_font=ImageFont.truetype(path+"Antonio-Regular.ttf",50)#设置水印字体类型和大小
    text_overlay=Image.new('RGBA',layer.size,(255,255,255,0))#生成与原图等大的透明图层
    #接下来画水印图层：
    image_draw=ImageDraw.Draw(text_overlay)
    text_size_x,text_size_y=image_draw.textsize(text,text_font)#获取水印大小
    
#     #正中间水印：
#     text_xy=(layer.size[0]//2-text_size_x//2, layer.size[1]//2-text_size_y//2)#获取水印位置，原点左上角
#     image_draw.text(text_xy,text,font=text_font,fill=(255,255,255,100))#设置水印颜色透明度位置
#     res1=Image.alpha_composite(layer,text_overlay)#叠加2图层
#     res1.save(savePath1+"\\"+str(i)+"_center.png")#保存
    
    #随机位置水印：
    pos_x=random.randint(0,layer.size[0]-text_size_x)
    pos_y=random.randint(0,layer.size[1]-text_size_y)
    text_xy=(int(pos_x), int(pos_y))#获取水印位置，原点左上角
    image_draw.text(text_xy,text,font=text_font,fill=(255,255,255,100))#设置水印颜色透明度位置
    res2=Image.alpha_composite(layer, text_overlay)#叠加2图层
    res2.save(savePath2+"\\"+str(i)+"_random.png")#保存
    print("success "+str(i))
    
```
